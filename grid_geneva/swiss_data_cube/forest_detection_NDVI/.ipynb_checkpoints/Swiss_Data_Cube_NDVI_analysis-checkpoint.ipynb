### PARAMETERS AND FUNCTIONS

```python
# Define the time window
year_splitting = [[1999,2000,2001,2002],[2003,2004,2005,2006,2007],[2008,2009,2010,2011,2012],[2013,2014,2015,2016,2017]]
#year_splitting = [[2013,2014,2015,2016,2017]]

start = str(year_splitting[0][0])+'-1-1'
end = str(year_splitting[-1][-1])+'-12-31'
start_date_init = datetime.strptime(start, '%Y-%m-%d')
end_date_init = datetime.strptime(end, '%Y-%m-%d')

plat ="LANDSAT_7"
prod = "ls7_ledaps_swiss"
#plat = "LANDSAT_5"
#prod = "ls5_ledaps_swiss"

## considered months
sel_months = [5,6,7,8,9]

## mask categories (2=cloud shadow, 3=snow, 4=cloud , 255=fill)
no_cats=[2,3,4,255]

## thresholds
ndvi_threshold=0.8
ts_percentage=0.45

## pixel resolution
resolution_km2=25*25/1000000

# dictionary for plotting later in the scatter plot
surface_dict={}
indicator_dict={}
```

```python
def tupple_to_cond(cats):
    for m in range(1,len(cats)+1):
        if m == 1: cond_cats = '(dataset_in.cf_mask == %i)' % (cats[m-1])
        else: cond_cats = '%s | (dataset_in.cf_mask == %i)' % (cond_cats, cats[m-1])
    return cond_cats
```

### CANTON GENEVA 

```python
# Canton GVA
min_lon = 5.93771 
max_lon = 6.31743
min_lat = 46.12089
max_lat = 46.37531
```

```python
## Canton GVA MASK
from osgeo import gdal
dataset = gdal.Open("GE_MASK.tif")

print("Number of layers ".format(dataset.RasterCount))
print("==== Some info ====")
print("Size of tif file is {} x {} x {}".format(dataset.RasterXSize,
                                    dataset.RasterYSize,
                                    dataset.RasterCount))
print("Projection is {}".format(dataset.GetProjection()))
mask_GE = np.array(dataset.GetRasterBand(1).ReadAsArray())
print("The shape of the array is {} ".format(mask_GE.shape))
```

### MINI DATA CUBE ANALYSIS

```python
tmp_dc = DataAccessApi()
acquisitions_list_init=tmp_dc.list_acquisition_dates(product=prod, platform=plat, 
        latitude=(min_lat, max_lat), longitude=(min_lon,max_lon),  
        time=(start_date_init,end_date_init))  
```

```python
## The LOOP

for row in year_splitting:   
    print(row)
    acquisitions_list_year = []
    for y in row:
        #print(y)
        for t in acquisitions_list_init:
            if (t.year==y): 
                acquisitions_list_year.append(t)
    #print("Acquisition list ",acquisitions_list_year)
    #print("====================")
    dataset_in = dc.load(platform=plat,
                     product=prod,
                     time=(acquisitions_list_year[0],acquisitions_list_year[-1]),
                     lon=(min_lon, max_lon), 
                     lat=(min_lat, max_lat),
                     measurements=['red', 'nir','cf_mask'])
    with np.errstate(invalid='ignore'):
        ndvi = (dataset_in.nir - dataset_in.red) / (dataset_in.nir + dataset_in.red)
    dataset_in['ndvi'] = ndvi.astype(float)

    with np.errstate(invalid='ignore'):
        select_ndvi = dataset_in.ndvi> ndvi_threshold
    dataset_in['forest_mask'] = select_ndvi.astype(int) 
    
    ## filtering for the months and the mask
    dataset_in = dataset_in.sel(time=np.isin(dataset_in['time.month'], sel_months))
    dataset_in = dataset_in.where(mask_GE==1)
    
    unfiltered_forest=dataset_in.forest_mask

    # create dataset for forests (ndvi>xx)
    unfiltered_forest_dataset = unfiltered_forest.astype(np.int16).to_dataset(name = "forest")

    if len(no_cats) > 0:
        no_data = eval(tupple_to_cond(no_cats))
        no_data = no_data.values ## create a mask for the pixels where we don't have clear sky (cf_mask = 2,3,4,255)
        # Apply no_data on MDC
        unfiltered_forest_dataset.forest.values[no_data] = 255 ## N.B. this sets all the values of forest where we 
        # have no data to 255. It is needed for the perform_timeseries_analysis step

    ts = perform_timeseries_analysis(unfiltered_forest_dataset,"forest", no_data = 255)
    
    # select normalized threshold
    with np.errstate(invalid='ignore'):
        forest = ts.normalized_data>ts_percentage
    dataset_in['final_forest'] = forest.astype(int)
    #print(dataset_in.final_forest)
    dataset_in.final_forest.plot(cmap='OrRd',figsize=(8, 6))
    
    tot_pixel_forest_tmp = np.count_nonzero(dataset_in.final_forest)
    #print(baseline_mosaic_tmp)
    #utilities.nan_to_num(forest_baseline_masked, 0)
    
    ## ========= Save the SDC dataset with results in netcdf file
    file_name = 'SDC_forest_%d_%d.nc' % (row[0],row[-1])
    dataset_in.final_forest.to_netcdf(file_name)
    ## ===========================================
    
    surface_tmp = tot_pixel_forest_tmp*resolution_km2
    surface_tot = np.count_nonzero(mask_GE==1)*resolution_km2
    print("The forest coverage is ", surface_tmp, "km2, for a total of",tot_pixel_forest_tmp,"/",np.count_nonzero(mask_GE==1)," pixels")
    print("The total surface ", surface_tot,"km2")
    print("===================================================")
    surface_dict['%d-%d' % (row[0],row[-1])]=surface_tmp
    indicator_dict['%d-%d' % (row[0],row[-1])]=surface_tmp/surface_tot
```

### ACCURACY (CANTON GENEVA)

```python
## Confronting 2018 forest observation from SITG with latest SDC prediction (2013-2017)

## === Accessing and stacking predictions ===
sdc_forest_dataset = xr.open_dataset('SDC_forest_2013_2017.nc')
sdc_forest_dataset_stacked = dataset_in.final_forest.stack(z=('latitude', 'longitude')).astype(np.int16).to_dataset(name = "SDC_forest_stacked")

## === Accessing and stacking true observations (SITG Geneva) ===
#sitg_GE = gdal.Open("FFP_CADASTRE_FORET_4326.tif") 
sitg_GE = gdal.Open("FFP_CADASTRE_FORET_4326_alltouched.tif") 
sitg_GE_array = xr.DataArray(sitg_GE.GetRasterBand(1).ReadAsArray(),
                                coords={'latitude': dataset_in.latitude,'longitude': dataset_in.longitude},
                                dims=['latitude','longitude'], 
                                name='forest_sitg')
sdc_forest_dataset_stacked['SITG_forest_stacked'] = sitg_GE_array.stack(z=('latitude', 'longitude')).astype(np.int16)

## === Transforming the mask of GVA canton into DataArray and stacking ===
mask_GE_DataArray = xr.DataArray(mask_GE,
                                coords={'latitude': dataset_in.latitude,'longitude': dataset_in.longitude},
                                dims=['latitude','longitude'], name='mask_GE')
stacked_mask_GE_DataArray = mask_GE_DataArray.stack(z=('latitude', 'longitude'))

## === Filtering the dataset with the stacked mask and creating the prediction and observation arrays ===
sdc_forest_dataset_filtered = sdc_forest_dataset_stacked.where(stacked_mask_GE_DataArray==1,drop=True)
prediction = sdc_forest_dataset_filtered.SDC_forest_stacked
observation = sdc_forest_dataset_filtered.SITG_forest_stacked
```

```python
# Accuracy, confusion matrix and all that
import itertools
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

def pretty_print_conf_matrix(y_true, y_pred, 
                             classes,
                             normalize=False,
                             text=False,
                             title='Confusion matrix',
                             cmap=plt.cm.Blues):
    """
    Mostly stolen from: http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py

    Normalization changed, classification_report stats added below plot
    """

    cm = confusion_matrix(y_true, y_pred)

    # Configure Confusion Matrix Plot Aesthetics (no text yet) 
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    #plt.title(title, fontsize=12) ## I don't need the title for the moment
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, fontsize=14, rotation=15)
    plt.yticks(tick_marks, classes, fontsize=14)
    plt.ylabel('True pixel class', fontsize=14)
    plt.xlabel('Predicted pixel class', fontsize=14)

    # Calculate normalized values (so all cells sum to 1) if desired
    if normalize:
        cm = np.round(cm.astype('float') / cm.sum(),3) #(axis=1)[:, np.newaxis]
    

    # Place Numbers as Text on Confusion Matrix Plot
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black",
                fontsize=14)

    if text: 
        # Add Precision, Recall, F-1 Score as Captions Below Plot
        rpt = classification_report(y_true, y_pred)
        rpt = rpt.replace('avg / total', '      avg')
        rpt = rpt.replace('support', 'N Obs')

        plt.annotate(rpt, 
                     xy = (0,0), 
                     xytext = (-70, -180), 
                     xycoords='axes fraction', textcoords='offset points',
                     fontsize=14, ha='left')    

    # Plot
    plt.tight_layout()
    plt.savefig('confusion_matrix.pdf')
```

```python
# Plot the confusion matrix

plt.style.use('classic')
plt.figure(figsize=(6,6))
pretty_print_conf_matrix(observation, prediction, 
                         classes= ['non-forest', 'forest'],
                         normalize=True, 
                         text=False,
                         title='Confusion Matrix')

## Change normalize=False to have the raw numbers instead of percentage in the plot and text=True if you are looking for the complete analysis
```
